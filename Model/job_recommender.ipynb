```python
from sentence_transformers import SentenceTransformer

model = SentenceTransformer('all-MiniLM-L6-v2')
```

```python
# Example data
user_profile = "Experienced software developer with expertise in Python and machine learning."
job_descriptions = [
    "Looking for a Python developer to join our team.",
    "Data scientist needed with experience in machine learning.",
    "Marketing manager with 5+ years of experience."
]

# Encoding
user_vector = model.encode(user_profile)
job_vectors = model.encode(job_descriptions)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarities = cosine_similarity([user_vector], job_vectors)
```

```python
import numpy as np

# Example: Recommend top 2 jobs
top_n_indices = np.argsort(similarities[0])[-2:]
recommended_jobs = [job_descriptions[i] for i in top_n_indices]

print("Recommended Jobs:", recommended_jobs)
```
